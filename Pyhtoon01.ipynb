### Reloading and Cleaning Data

First, let's reload the original data, drop the entirely null 'Status' and 'unnamed1' columns, and handle the remaining missing values in 'Amount' before converting its datatype.

```python
df = pd.read_csv("/content/Diwali Sales Data.csv", encoding="unicode_escape")

# Drop 'Status' and 'unnamed1' columns which are entirely null
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)

# Drop rows where 'Amount' is NaN (as there are only a few)
df.dropna(subset=['Amount'], inplace=True)

# Convert 'Amount' column to integer type
df['Amount'] = df['Amount'].astype(int)

display(df.info())
```

### Creating Countplot for Gender

Now that the data is cleaned and loaded, we can create the countplot for 'Gender'.

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = sns.countplot(x='Gender', data=df)

# Add labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)

plt.title('Count of Orders by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Orders')
plt.show()
```

### Analyzing Amount Spent by Gender

Let's visualize the total amount spent by each gender to understand their contribution to sales.

```python
sales_by_gender = df.groupby('Gender')['Amount'].sum().reset_index()

sns.barplot(x='Gender', y='Amount', data=sales_by_gender, hue='Gender', palette='viridis', legend=False)

plt.title('Total Amount Spent by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.show()
```

```python
grouped_data = df.groupby('Gender', as_index=False).sum()
final_result = grouped_data.sort_values(by='Amount', ascending=False)
display(final_result)
```

```python
sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data=sales_gen, hue='Gender', palette='viridis', legend=False)

plt.title('Total Amount Spent by Gender (Sorted)')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.show()
```
